## 8. For the given input values of x and y, build a linear regression model using scipy  library and plot it.  

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Training data (x and y values)
x_train = [3, 8, 9, 13, 3, 6, 11, 21]
y_train = [30, 57, 64, 72, 36, 43, 59, 90]
```

```python
# Perform linear regression using scipy's linregress function
slope, intercept, r_value, p_value, std_err = stats.linregress(x_train, y_train)

# Print out the regression parameters
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)
print("P-value:", p_value)
print("Standard Error:", std_err)

# Create the regression line (y = mx + b)
regression_line = [slope * x + intercept for x in x_train]

# Plot the original data and the regression line
plt.scatter(x_train, y_train, color='blue', label="Data Points")
plt.plot(x_train, regression_line, color='red', label="Regression Line")
plt.title("Linear Regression with scipy")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
