```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('buyscomputer.csv')
```

```python
data
```

```python
le = LabelEncoder()
data['age'] = le.fit_transform(data['age'])
data['income'] = le.fit_transform(data['income'])
data['student'] = le.fit_transform(data['student'])
data['credit_rating'] = le.fit_transform(data['credit_rating'])
data['buys_computer'] = le.fit_transform(data['buys_computer'])
data
```

```python
x_train = data.iloc[:,:-1]
y_train = data.iloc[:,-1]
```

```python
x_train
```

```python
y_train
```

```python
model = DecisionTreeClassifier()
model = model.fit(x_train,y_train)
```

```python
tree.plot_tree(model, feature_names=x_train.columns, class_names=['No', 'Yes'], filled=True)
plt.show()
```
