### 7. For the given input values of x and y, compute coefficient values to build linear  regression model and plot it. Use this model to predict the value of y for the new  value of x. Also compute MSE and RSS. (Note: Use only numpy library) 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class LinearRegression:
  def __init__(self, x, y):
    self.x = np.array(x)
    self.y = np.array(y)

  def fit(self):
    mean_x = self.x.mean()
    mean_y = self.y.mean()

    # Compute coefficients
    numerator = ((self.x - mean_x) * (self.y - mean_y)).sum()
    denominator = ((self.x - mean_x) ** 2).sum()

    self.B1 = numerator / denominator
    self.B0 = mean_y - self.B1 * mean_x

  def predict(self, x):
    X=np.array(x)
    return self.B0 + self.B1 * X

  def accuracy_score(self, y_true, y_pred):
    # Compute RSS (Residual Sum of Squares)
        rss = ((y_true - y_pred) ** 2).sum()
        # Compute MSE (Mean Squared Error)
        mse = rss / len(y_true)

        print("RSS (Residual Sum of Squares):", rss)
        print("MSE (Mean Squared Error):", mse)
```

```python
# Training data
x_train = [3, 8, 9, 13, 3, 6, 11]
y_train = [30, 57, 64, 72, 36, 43, 59]

# Testing data
x_test = [1, 16, 21]
y_test = [20, 83, 90]

def main(x_train, y_train, x_test, y_test):
  # Visualize training data
  plt.scatter(x_train, y_train, label = "Data Points")
  plt.title("Training Data")
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.legend()
  plt.show()

  # Visualize testing data
  plt.scatter(x_test, y_test,label = "Data Points")
  plt.title("Testing Data")
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.legend()
  plt.show()

  # Create and train the model
  model = LinearRegression(x_train, y_train)
  model.fit()

  # Predict on training data
  train_predictions = model.predict(x_train)
  # Visualize the training data along with the regression line
  plt.scatter(x_train, y_train)
  plt.plot(x_train, train_predictions, c="red", label="Regression Line")
  plt.title("Linear Regression Model (Training Data)")
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.legend()
  plt.show()
  
  # Predict on test data
  test_predictions = model.predict(x_test)
  # Visualize predictions on test data
  plt.scatter(x_test, y_test, label='Actual Testing Data', c="blue")
  plt.plot(x_test, test_predictions, c="green", label="Predicted Testing Line")
  plt.title("Test Data Predictions")
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.legend()
  plt.show()

  # Display predicted results
  print("Predictions on Training Data:", train_predictions)
  print("Predictions on Test Data:", test_predictions)

  # Evaluate model using MSE and RSS
  model.accuracy_score(y_test, test_predictions)


  # Final comparison between actual test data and predicted values
  plt.figure(figsize=(10, 5))
  plt.scatter(x_train, y_train, label='Training Data', c="red")
  plt.scatter(x_test, test_predictions, label='Predicted Testing Data', c="green")
  plt.plot(x_test, test_predictions, c="green", label="Predicted Testing Line")
  plt.scatter(x_test, y_test, label='Actual Testing Data', c="blue")
  plt.title("Comparison: Actual vs Predicted")
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.legend()
  plt.show()

# Run the main function
if __name__ == "__main__":
    main(x_train, y_train, x_test, y_test)
    # This ensures that main() only runs if you execute that file directly, not when you import it.
```
