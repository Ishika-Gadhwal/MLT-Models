## 11. Build a linear regression model using SKlearn library for the following :
### a. Take input values in terms of lists (X,Y).
### b. Generate training dataset and testing dataset from the input lists using train_test_split() from SKlearn.
### c. Train the model using training dataset.
### d. Predict Y for test dataset.
### e. Draw the plot.
### f. Repeat step (b)-(e) for random pairs of (X,Y).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x,y = make_regression(n_samples=100,n_features=1,noise=11)
# x = x.flatten()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=43)
print(x_train.size)
x_test.size
```

```python
# Visualize training data
plt.scatter(x_train, y_train)
plt.title("Training Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Visualize testing data
plt.scatter(x_test, y_test)
plt.title("Testing Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred_trainingData = model.predict(x_train)
```

```python
# Visualize the training data along with the regression line
plt.scatter(x_train, y_train)
plt.plot(x_train, y_pred_trainingData, c="red", label="Regression Line")
plt.title("Linear Regression Model (Training Data)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
y_pred = model.predict(x_test)
```

```python
# Visualize predictions on test data
plt.scatter(x_test, y_test, label='Actual Testing Data', c="blue")
plt.plot(x_test,y_pred, c="green", label="Predicted Testing Line")
plt.title("Test Data Predictions")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Final comparison between actual test data and predicted values
plt.figure(figsize=(10, 5))
plt.scatter(x_train, y_train, label='Training Data', c="red")
plt.scatter(x_test, y_pred, label='Predicted Testing Data', c="green")
plt.plot(x_test, y_pred, c="green", label="Regression Line")
plt.scatter(x_test, y_test, label='Actual Testing Data', c="blue")
plt.title('Linear Regression Fit')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
r2_score(y_test,y_pred) # Close to 1 means 100% accuracy
```

```python
mean_squared_error(y_test,y_pred)
```
